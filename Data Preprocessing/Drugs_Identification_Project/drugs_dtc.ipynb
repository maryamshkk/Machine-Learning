Data Preprocessing 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train = df.drop(['Drug'], axis=1)
y_train = df['Drug']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)
```

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
sex_col = ['Sex']
bp_col = ['BP']
chol_col = ['Cholesterol']
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('sex_tnf', OneHotEncoder(drop='first'), sex_cols),
    ('bp_tnf', OrdinalEncoder(categories=[['LOW', 'NORMAL', 'HIGH']]), bp_cols),
    ('chol_tnf', OrdinalEncoder(categories=[['NORMAL', 'HIGH']]), chol_col)
], remainder='passthrough')
```

```python
X_train_new = preprocessor.fit_transform(X_train)
X_test_new = preprocessor.transform(X_test)
```

```python
print(" Data encoding done successfully!")
print("Training shape:", X_train_new.shape)
print("Testing shape:", X_test_new.shape)
```

```python
pd.DataFrame(X_train_new, columns=preprocessor.get_feature_names_out())
```

Training Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(criterion='entropy', max_depth=4, random_state=42)
```

```python
dtc.fit(X_train_new, y_train)
```

```python
print("Model trained successfully!")
print("Tree Depth:", dtc.get_depth())
print("Number of Leaves:", dtc.get_n_leaves())
```

Testing 

```python
y_pred = dtc.predict(X_test_new)
```

Evaluation 

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Accuracy metrix
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy Score', round(accuracy, 2))
```

```python
# Confusion metrix
cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix', cm)
```

```python
# Classification Report
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

Visualization 

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(15, 10))
plot_tree(
    dtc, 
    filled=True,
    feature_names=preprocessor.get_feature_names_out(),
    class_names=y_train.unique(),
    rounded=True,
    fontsize=10
)
plt.title('Decision Tree Visualization from Drug Classification', fontsize=16)
plt.show()
```
