```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
# Data
X = np.array([
    [1,5],
    [2,6],
    [3,5],
    [4,4],
    [5,3]
], dtype=float)

y = np.array([50, 60, 65, 70, 75], dtype=float)
```

```python
# Create and train model 
model = LinearRegression()
model.fit(X, y)
```

```python
# Coefficients
print("Intercept (b0)", model.intercept_)
print("Slopes (b1, b2)", model.coef_)
```

```python
X_pred = np.array([[6,4]])
y_pred = model.predict(X_pred)
print("Predicted score:", y_pred[0])
```

```python
# Model evaluation
y_pred_all = model.predict(X)
r2 = r2_score(y, y_pred_all)
mse = mean_squared_error(y, y_pred_all)
mae = mean_absolute_error(y, y_pred_all)
print(f"R square Score, {r2:.4f}")
print(f"MSE:  {mse:.4f}")
print(f"MAE: , {mae:.4f}")
```

```python
# 3d plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:, 0], X[:, 1], y, color='red', label='Actual Data')

x_surf, y_surf = np.meshgrid(
    np.linspace(X[:, 0].min(), X[:, 0].max(), 10),
    np.linspace(X[:, 1].min(), X[:, 1].max(), 10)
)
z_surf = model.intercept_ + model.coef_[0] * x_surf + model.coef_[1] * y_surf

ax.plot_surface(x_surf, y_surf, z_surf, alpha=0.5, cmap='viridis')

ax.set_xlabel('Hours Studied')
ax.set_ylabel('Hours Slept')
ax.set_zlabel('Exam Score')
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
actual = np.array([10, 15, 20, 25, 30])
predicted = np.array([12, 14, 22, 23, 29])

# Create scatter plot
plt.figure(figsize=(8,6))
plt.scatter(actual, predicted, color='blue', label='Data Points')

# Add diagonal line for perfect predictions
plt.plot([actual.min(), actual.max()], [actual.min(), actual.max()], 'r--', label='Perfect Fit')

# Annotate each point with actual & predicted values
for i in range(len(actual)):
    plt.annotate(f"A:{actual[i]}, P:{predicted[i]}",
                 (actual[i], predicted[i]),
                 textcoords="offset points", xytext=(5,5),
                 fontsize=8, color="black")

plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted (Annotated)")
plt.legend()
plt.grid(True)
plt.show()
```
