```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
# Data
x = np.array([
    [3.5, 2.0, 10.5, 4, 120],
    [3.69, 2.2, 11.0, 6, 150],
    [3.44, 1.8, 9.8, 4, 110],
    [3.43, 2.1, 11.5, 6, 130],
    [4.34, 3.0, 12.0, 8, 180],
    [4.42, 3.1, 12.2, 8, 190],
    [2.37, 1.6, 9.0, 4, 100]
])
y = np.array([18, 15, 18, 16, 15, 14, 24])
```

```python
# Model 
model = LinearRegression()
model.fit(x,y)
```

```python
x_pred = np.array([[4.0, 2.5, 11.0, 6, 140]])
y_pred = model.predict(x_pred)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data showing a negative correlation
np.random.seed(42)  # For reproducibility

# Engine displacement between 1000cc and 5000cc
displacement = np.random.uniform(1000, 5000, 100)

# MPG decreases as displacement increases, with some noise
mpg = 50 - (displacement / 100) + np.random.normal(0, 3, 100)  # Noise added

# Plot
plt.scatter(displacement, mpg, color='blue', alpha=0.7, label='Car Data Points')
plt.xlabel("Engine Displacement (cc)")
plt.ylabel("Miles per Gallon (MPG)")
plt.title("Engine Displacement vs MPG (Negative Correlation)")
plt.grid(True)
plt.legend()
plt.show()
```
