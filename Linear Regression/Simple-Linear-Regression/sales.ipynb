# Linear Regression with Scikit-learn

This notebook demonstrates how to perform simple linear regression using `scikit-learn` and visualizes the predicted value with a green point on the regression line.

Imports 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

Data 

```python
X = np.array([1,2,3,4,5])
y = np.array([2,4,5,4.5,6])
```

Reshape 

```python
X = X.reshape(-1,1)     # 2d array
```

Create and Train Model 

```python
model = LinearRegression()
model.fit(X, y)
```

Predict for X and new value

```python
y_pred = model.predict(X)
X_new = np.array([[6]]) 
y_new = model.predict(X_new)
```

Equation and Prediction

```python
print("Slope", model.coef_[0])
print("Intercept", model.intercept_)
print(f"Equation : Y={model.coef_[0]:.2f}X + {model.intercept_:.2f}")
print(f"Prediction for X=6: {y_new[0]:.2f}")
```

Plot 

```python
plt.figure(figsize=(10,4))
plt.scatter(X, y, label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.scatter(X_new, y_new, color='green', label=f'Prediction X=6, Y={y_new[0]:.2f}' )
# For visualizing greeen point
plt.axvline(x=6, color='green', linestyle='dotted')
plt.axhline(y=y_new[0], color='green', linestyle='dotted')

plt.xlabel('Advertising Sales')
plt.ylabel('Sales')
plt.title('Linear Regression using Scikit-learn')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("Mean Squared Error", mean_squared_error(y, y_pred))
print("R2 Value", r2_score(y, y_pred))
```
