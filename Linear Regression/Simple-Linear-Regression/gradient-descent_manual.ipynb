```python
import numpy as np 
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 
```

```python
X = np.array([1,2,3,4,5], dtype=float)
y = np.array([2,4,5,4,5], dtype=float)

n = len(X)
```

```python
m = 0.0     #slope
b = 0.0     #intercept
alpha = 0.01        #learning rate
epochs = 1000
```

```python
for _ in range(epochs):
    y_pred = m*X +b
    
    # calculate gradients
    dm = (-2/n) * sum(X * (y-y_pred))
    db = (-2/n) * sum(y - y_pred)
    
    # update m and b
    m = m - alpha * dm
    b = b - alpha * db 
```

```python
print(f"Final Slope (m): {m}")
print(f"Final Intercept (c): {b}")
```

```python
hours = 6 
predicted_score = m * hours + b
print(f"Predicted score for {hours} hours studied: {predicted_score}")
```
