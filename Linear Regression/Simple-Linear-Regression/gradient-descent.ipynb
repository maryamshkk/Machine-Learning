```python
import pandas as pd 
import numpy as np 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 
```

```python
X = np.array([1,2,3,4,5], dtype = float)
y = np.array([2,4,5,4,5], dtype=float)
```

```python
# Gradient Descent Implementation
n = len(X)
m = 0.0     #slope
b = 0.0     #intercept
alpha = 0.01        #learning rate
epochs = 1000
```

```python
for _ in range(epochs):
    y_pred = m*X +b
    
    # calculate gradients
    dm = (-2/n) * sum(X * (y-y_pred))
    db = (-2/n) * sum(y - y_pred)
    
    # update m and b
    m = m - alpha * dm
    b = b - alpha * db 
    
y_pred_gd = m * X + b 
```

```python
print(f"Final Slope (m): {m}")
print(f"Final Intercept (c): {b}")
print(f"Prediction for 6 hours studied: {m*6 + b}")
```

```python
# SKlearn Implementation
X_reshaped = X.reshape(-1,1)
model = LinearRegression()
model.fit(X_reshaped, y)
```

```python
m_sklearn = model.coef_[0]
b_sklearn = model.intercept_
y_pred_sklearn = model.predict(X_reshaped)
```

```python

print("\n===== Scikit-Learn Results =====")
print(f"Slope (m): {m_sklearn}")
print(f"Intercept (b): {b_sklearn}")
print(f"Prediction for 6 hours studied: {model.predict([[6]])[0]}")
```

```python
plt.scatter(X, y, color='red', label="Actual Data")  # Data points
plt.plot(X, y_pred_gd, color='blue', linestyle='--', label='Gradient Descent Line')  # GD line
plt.plot(X, y_pred_sklearn, color='green', label='Scikit-Learn Line')  # sklearn line
plt.xlabel('Hours studied')  # X-axis label
plt.ylabel('Marks Scored')   # Y-axis label
plt.legend()                 # Show labels
plt.show()                   # Display plot
```
