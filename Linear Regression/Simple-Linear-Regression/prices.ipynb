Imports

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

Data

```python
data = {'SquareFootage': [800, 1000, 1200, 1500, 1800],
        'Price': [200, 250, 300, 350, 400]}
df = pd.DataFrame(data)

X =  df['SquareFootage']
y = df['Price']
```

Reshape X for sklearn

```python
X = X.reshape(-1, 1)
```

Create and Train Model

```python
model = LinearRegression()
model.fit(X, y)
```

```python
print(f"Regression Equation: Price = {model.coef_[0]:.2f} * SquareFootage + {model.intercept_:.2f}")
```

Prediction

```python
y_pred = model.predict(X)
X_new = np.array([[1600]])
y_new = model.predict(X_new)
```

Evaluation and Prediction

```python
print(f"Equation: Y={model.coef_[0]:.2f}X + {model.intercept_:.2f}")
print(f"Prediction for X=1600 {y_new[0]:.2f}")
```

Plot

```python
plt.figure(figsize=(10,4))
plt.scatter(X, y, label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.scatter(X_new, y_new, color='green', label=f'Prediction (X=1600, Y={y_new[0]:.2f})')
plt.xlabel('Square Footage', fontsize = 12)
plt.ylabel('Price ($1000)', fontsize = 12)
plt.title('Linear Regression using Scikit-Learn')
plt.legend()
plt.grid(True)
plt.show()
```

Evaluation

```python
print("Mean Squared Error (MSE):", mean_squared_error(y, y_pred))
print("R2 Score:", r2_score(y, y_pred))
```
