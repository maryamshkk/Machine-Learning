```python
# Imports
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Data
X = [1],[2],[3]
y = [2,4,6]
```

```python
# Model Training
model = LinearRegression()
model.fit(X,y)
```

```python
# Prediction
x_pred = [[4]]
y_pred = model.predict(x_pred)
```

```python
# PLot
plt.scatter(X,y, color='blue', label='Training data')
plt.plot([1,2,3,4], model.predict([[1],[2],[3],[4]]), color='red', label='regression Line')
plt.scatter(x_pred,y_pred, color='green', label='Prediction (4,{:.2f})'.format(y_pred[0]))

plt.title('Linear Regression Prediction')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid('True')
plt.show()
```
