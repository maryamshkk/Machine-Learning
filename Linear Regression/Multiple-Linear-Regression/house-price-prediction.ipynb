Imports 

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

Data 

```python
# X1 = Area, X2 = Bedrooms
X = np.array([
    [1000, 2],
    [1500, 3],
    [2000, 3],
    [2500, 4],
    [3000, 4]
])
#  Target: Price in $1000s
y = np.array([200, 300, 350, 400, 500])
```

Create and Train Model 

```python
model = LinearRegression()
model.fit(X, y)
```

Prediction

```python
y_pred = model.predict(X)
```

Coefficients and Intercept

```python
# Slope(m1,m2) and Intercept(c)
print("Coefficients (m1, m2)", model.coef_)
print("Intercept (c)", model.intercept_)
```

Predict a new value 

```python
X_new = np.array([[2700, 3]])
y_new = model.predict(X_new)
print(f"Prediction for 2700 sqft, 3 bedrooms: ${y_new[0]:.2f}")
```

Plot 

```python
plt.figure(figsize=(12, 4))
plt.plot(y, label="Actual Price", marker='o')
plt.plot(y_pred, label="Predicted Price", marker='x')
plt.title("Actual vs Predicted Prices")
plt.xlabel("Data Point Index")
plt.ylabel("Price ($1000)")
plt.legend()
plt.grid(True)
plt.show()
```
